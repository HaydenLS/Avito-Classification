Наблюдается дисбаланс классов. Рассмотрим распределение по категориям.

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Распределение цветов по категориям', fontsize=16)

for i, category in enumerate(df['category'].unique()):
    ax = axes[i//2, i%2]  # Выбор позиции в сетке (0,0), (0,1), (1,0), (1,1)
    
    
    subset = df[df['category'] == category]
    color_counts = subset['target'].value_counts()

    
    color_counts.plot(kind='bar', ax=ax, edgecolor='black', alpha=0.7)
    
    # Настройки графика
    ax.set_title(f'Категория: {category}')
    ax.set_xlabel('Цвет')
    ax.set_ylabel('Количество')
    ax.grid(axis='y', linestyle='--', alpha=0.6)

# Автоматическая регулировка отступов
plt.tight_layout()
plt.show()
```
